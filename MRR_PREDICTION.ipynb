# MRR PREDICTION USING LINEAR REGRESSION MODEL 
This project involves the use of a linear regression model to understand and predict user growth and MRR, future revenue, conversion, churn, and more. The model is trained using a sample data of 50 rows, which includes the number of users and the corresponding revenue. The model is then used to predict the revenue for 2000 users. The results are visualized using a scatter plot and a regression line.

This project can drive business decisions by providing insights into the relationship between the number of users and the revenue generated. The model can be used to make informed decisions about user acquisition and retention strategies based on the revenue generated. Additionally, the model can be used to identify trends in user growth and revenue over time, allowing the business to make informed decisions about future investments and growth strategies. Overall, the linear regression model is a powerful tool for analyzing user growth and revenue data and making informed business decisions based on that analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'users': [100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 
              200, 210, 220, 230, 240, 250, 260, 270, 280, 290, 
              300, 310, 320, 330, 340, 350, 360, 370, 380, 390, 
              400, 410, 420, 430, 440, 450, 460, 470, 480, 490,
              500, 510, 520, 530, 540, 550, 560, 570, 580, 590],
    'revenue': [10000 + np.random.normal(0, 2000), 11000 + np.random.normal(0, 2000), 12000 + np.random.normal(0, 2000), 13000 + np.random.normal(0, 2000), 14000 + np.random.normal(0, 2000), 
                 15000 + np.random.normal(0, 2000), 16000 + np.random.normal(0, 2000), 17000 + np.random.normal(0, 2000), 18000 + np.random.normal(0, 2000), 19000 + np.random.normal(0, 2000),
                 20000 + np.random.normal(0, 2000), 21000 + np.random.normal(0, 2000), 22000 + np.random.normal(0, 2000), 23000 + np.random.normal(0, 2000), 24000 + np.random.normal(0, 2000),
                 25000 + np.random.normal(0, 2000), 26000 + np.random.normal(0, 2000), 27000 + np.random.normal(0, 2000), 28000 + np.random.normal(0, 2000), 29000 + np.random.normal(0, 2000),
                 30000 + np.random.normal(0, 2000), 31000 + np.random.normal(0, 2000), 32000 + np.random.normal(0, 2000), 33000 + np.random.normal(0, 2000), 34000 + np.random.normal(0, 2000),
                35000 + np.random.normal(0, 2000), 36000 + np.random.normal(0, 2000), 37000 + np.random.normal(0, 2000), 38000 + np.random.normal(0, 2000), 39000 + np.random.normal(0, 2000),
                40000 + np.random.normal(0, 2000), 41000 + np.random.normal(0, 2000), 42000 + np.random.normal(0, 2000), 43000 + np.random.normal(0, 2000), 44000 + np.random.normal(0, 2000),
                45000 + np.random.normal(0, 2000), 46000 + np.random.normal(0, 2000), 47000 + np.random.normal(0, 2000), 48000 + np.random.normal(0, 2000), 49000 + np.random.normal(0, 2000),
                50000 + np.random.normal(0, 2000), 51000 + np.random.normal(0, 2000), 52000 + np.random.normal(0, 2000), 53000 + np.random.normal(0, 2000), 54000 + np.random.normal(0, 2000),
                55000 + np.random.normal(0, 2000), 56000 + np.random.normal(0, 2000), 57000 + np.random.normal(0, 2000), 58000 + np.random.normal(0, 2000), 59000 + np.random.normal(0, 2000)]}

# Convert data to pandas dataframe
df = pd.DataFrame(data)
```

Split data into training and test sets

```python
X_train = df[['users']]
y_train = df['revenue']
```

Fit the linear regression model

```python
reg = LinearRegression().fit(X_train, y_train)
```

Predict revenue for 2000 users

```python
X_test = [[2000]]
y_pred = reg.predict(X_test)
print("Revenue for 2000 users:", y_pred[0])
```

Plot the training data and regression line

```python
plt.scatter(X_train, y_train, color='blue')
plt.plot(X_train, reg.predict(X_train), color='red', linewidth=2)
plt.xlabel('Users')
plt.ylabel('Revenue')
plt.show()
```
