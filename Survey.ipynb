```python
#Question 4
#What % of people from each country that knows how to use python

country_respondents=df['Country'].value_counts()
country_respondents
```

```python
country_uses_python=country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
country_uses_python
```

```python
python_df=pd.concat([country_respondents, country_uses_python], axis='columns', sort=False).head()
python_df
```

```python
#lets rename the column 'LanguageHaveWorkedWith' to 'numknowpython'

python_df.rename(columns={'LanguageHaveWorkedWith' : 'numknowpython', 'Country':'numrespondents'}, inplace=True)
```

```python
python_df
```

```python
#to calculate for % that knows python in each county and add it to the dataframe as a new column

python_df['percknowpython']=(python_df['numknowpython']/python_df['numrespondents'])*100
python_df
```

```python
#to sort them by the highest percentage

python_df.sort_values(by='percknowpython', ascending=False, inplace=True)
python_df.head(50)
```
