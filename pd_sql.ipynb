## Connect Pandas with postgres and query the table using python

```python
from sqlalchemy import create_engine
import pandas as pd

# create a connection to the database
engine = create_engine('postgresql://postgres:DBpassword@localhost:5432/DEMO_DB')

# retrieve data from the database and store it in a DataFrame
df = pd.read_sql_query("SELECT * FROM \"COVID\"", engine)


# sort the DataFrame by a specific column in descending order
df.sort_values(by='year', ascending=True, inplace=True)
```

```python
#using query
df_DE = df.query('country == "Germany" & year >= 2020')
df_DE
```

```python
# Usinf loc
# filter the DataFrame by column 'Country' and value 'Germany', and column 'Year' and value >2010
df_DE = df.loc[(df['country'] == 'Germany') & (df['year'] >= 2020)]

#print the filtered data
print(df_DE)
```
